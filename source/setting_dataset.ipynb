```python
import os
import json
import numpy as np
from tqdm import tqdm
from collections import OrderedDict
```

```python
clinical_data_path = '/inye/dataset/clinical_data.json'
with open(clinical_data_path, 'r') as f:
    clinical_data = json.load(f)

data_split_path = '/inye/dataset/data_split.json'
with open(data_split_path, 'r') as f:
    data_split = json.load(f)
```

```python
preImg_of_testset = []
for sample in data_split['test']:
    preImg_of_testset.append(clinical_data[sample]['Pre_img_name'])
```

```python
placeholder_tokens = [
    "<DX-CN>",
    "<DX-MCI>",
    "<DX-AD>",
    "<Age--64>",
    "<Age-65-74>",
    "<Age-75-84>",
    "<Age-85->",
    "<Gender-Male>",
    "<Gender-Female>",
    "<Edu-6-12>",
    "<Edu-13-16>",
    "<Edu-17-18>",
    "<Edu-19->",
    "<Race-White>",
    "<Race-Black>",
    "<Race-Asian>",
    "<Race-MoreThanOne>",
    "<Race-Unknown>",
    "<Race-Indian-Alaskan>",
    "<Race-Hawwaiian-otherPI>",
    "<Marry-Married>",
    "<Marry-Widowed>",
    "<Marry-Divorced>",
    "<Marry-NeverMarried>",
    "<Marry-Unknown>",
    "<APOE4-0>",
    "<APOE4-1>",
    "<APOE4-2>",
]
```

```python
img_root = '/inye/dataset/T1_2D_slice_512/'
imgs = os.listdir(img_root)
```

```python
trainset = []
testset = []
testset_preImg = []

for img in tqdm(clinical_data):
    if img+".nii.gz" not in imgs:
        print("Not found", img)
        continue
    dx = clinical_data[img]['DX']
    age = clinical_data[img]['age']
    gender = clinical_data[img]['gender']
    education = clinical_data[img]['educat']
    race = clinical_data[img]['race']
    marry = clinical_data[img]['marry']
    apoe = clinical_data[img]['APOE4']

    prompt = ""
    if dx == "CN":
        prompt += placeholder_tokens[0] + " "
    elif dx == "MCI":
        prompt += placeholder_tokens[1] + " "
    elif dx == "Dementia":
        prompt += placeholder_tokens[2] + " "
    elif dx is not None:
        print("DX", dx, img)

    if dx == "CN":
        prompt += placeholder_tokens[0] + " "
    elif dx == "MCI":
        prompt += placeholder_tokens[1] + " "
    elif dx == "Dementia":
        prompt += placeholder_tokens[2] + " "
    elif dx is not None:
        print("DX", dx, img)

    if dx == "CN":
        prompt += placeholder_tokens[0] + " "
    elif dx == "MCI":
        prompt += placeholder_tokens[1] + " "
    elif dx == "Dementia":
        prompt += placeholder_tokens[2] + " "
    elif dx is not None:
        print("DX", dx, img)

    if dx == "CN":
        prompt += placeholder_tokens[0] + " "
    elif dx == "MCI":
        prompt += placeholder_tokens[1] + " "
    elif dx == "Dementia":
        prompt += placeholder_tokens[2] + " "
    elif dx is not None:
        print("DX", dx, img)

    if dx == "CN":
        prompt += placeholder_tokens[0] + " "
    elif dx == "MCI":
        prompt += placeholder_tokens[1] + " "
    elif dx == "Dementia":
        prompt += placeholder_tokens[2] + " "
    elif dx is not None:
        print("DX", dx, img)

    if dx == "CN":
        prompt += placeholder_tokens[0] + " "
    elif dx == "MCI":
        prompt += placeholder_tokens[1] + " "
    elif dx == "Dementia":
        prompt += placeholder_tokens[2] + " "
    elif dx is not None:
        print("DX", dx, img)

    if dx == "CN":
        prompt += placeholder_tokens[0] + " "
    elif dx == "MCI":
        prompt += placeholder_tokens[1] + " "
    elif dx == "Dementia":
        prompt += placeholder_tokens[2] + " "
    elif dx is not None:
        print("DX", dx, img)

    if dx == "CN":
        prompt += placeholder_tokens[0] + " "
    elif dx == "MCI":
        prompt += placeholder_tokens[1] + " "
    elif dx == "Dementia":
        prompt += placeholder_tokens[2] + " "
    elif dx is not None:
        print("DX", dx, img)

    if dx == "CN":
        prompt += placeholder_tokens[0] + " "
    elif dx == "MCI":
        prompt += placeholder_tokens[1] + " "
    elif dx == "Dementia":
        prompt += placeholder_tokens[2] + " "
    elif dx is not None:
        print("DX", dx, img)

    if dx == "CN":
        prompt += placeholder_tokens[0] + " "
    elif dx == "MCI":
        prompt += placeholder_tokens[1] + " "
    elif dx == "Dementia":
        prompt += placeholder_tokens[2] + " "
    elif dx is not None:
        print("DX", dx, img)

######################################

    if age is not None:
        if age < 65:
            prompt += placeholder_tokens[3] + " "
        elif age >= 65 and age < 75:
            prompt += placeholder_tokens[4] + " "
        elif age >= 75 and age < 85:
            prompt += placeholder_tokens[5] + " "
        elif age >= 85:
            prompt += placeholder_tokens[6] + " "
        else:
            print("Age", age, img)

    if gender == 'Male':
        prompt += placeholder_tokens[7] + " "
    elif gender == 'Female':
        prompt += placeholder_tokens[8] + " "
    elif gender is not None:
        print("Gender", gender, img)
    
    if education is not None:
        if education >= 6 and education < 13:
            prompt += placeholder_tokens[9] + " "
        elif education >= 13 and education < 17:
            prompt += placeholder_tokens[10] + " "
        elif education >= 17 and education < 19:
            prompt += placeholder_tokens[11] + " "
        elif education >= 19:
            prompt += placeholder_tokens[12] + " "
        else:
            print("Edu", education, img)
    
    if race == 'White':
        prompt += placeholder_tokens[13] + " "
    elif race == 'Black':
        prompt += placeholder_tokens[14] + " "
    elif race == 'Asian':
        prompt += placeholder_tokens[15] + " "
    elif race == 'More than one':
        prompt += placeholder_tokens[16] + " "
    elif race == 'Unknown':
        prompt += placeholder_tokens[17] + " "
    elif race == 'Am Indian/Alaskan':
        prompt += placeholder_tokens[18] + " "
    elif race == 'Hawaiian/Other PI':
        prompt += placeholder_tokens[19] + " "
    elif race is not None:
        print("Race", race, img)

    if marry == 'Married':
        prompt += placeholder_tokens[20] + " "
    elif marry == 'Widowed':
        prompt += placeholder_tokens[21] + " "
    elif marry == 'Divorced':
        prompt += placeholder_tokens[22] + " "
    elif marry == 'Never married':
        prompt += placeholder_tokens[23] + " "
    elif marry == 'Unknown':
        prompt += placeholder_tokens[24] + " "
    elif marry is not None:
        print("Marry", marry, img)
    
    if apoe is not None:
        if apoe == 0:
            prompt += placeholder_tokens[25]
        elif apoe == 1:
            prompt += placeholder_tokens[26]
        elif apoe == 2:
            prompt += placeholder_tokens[27]
        else:
            print("APOE4", apoe, img)

    my_data = OrderedDict()
    my_data['img'] = img
    my_data['prompt'] = prompt
    my_data['DX'] = dx
    my_data['Age'] = age
    my_data['Gender'] = gender
    my_data['Edu'] = education
    my_data['Race'] = race
    my_data['Marry'] = marry
    my_data['APOE4'] = apoe
    my_data['preImg'] = clinical_data[img]['Pre_img_name']

    # if img in preImg_of_testset:
    #     testset_preImg.append(my_data)
    if img in data_split['train']:
        trainset.append(my_data)
    elif img in data_split['test']:
        testset.append(my_data)
```
